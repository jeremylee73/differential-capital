```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#data import
returns = pd.read_pickle("../Data/returns.pkl")
```

```python
#drop very first row
returns = returns.iloc[1:, :]
```

```python
#sample stock
stock = '905270'
```

```python
#time period
t = pd.to_datetime('1999-02-24')
```

```python
#save the target (sample stock's returns upto time t)
y = returns[[stock]].loc[:t, :]
y.columns = ['Outcome']
y = y.sort_index(ascending = False)
y
```

```python
#shift the features
X = returns.loc[:(t + pd.DateOffset(-1)), :]
X = X.sort_index(ascending = False)
X
```

```python
#concatenate outcome and features
df_full = pd.concat([X, y], axis = 1)
df_full = df_full.sort_index(ascending = False)
df_full
```

```python
#shit outcome down 1 row
df_full['Outcome'] = df_full['Outcome'].shift(1)
df_full = df_full.iloc[1:, :]
df_full
```

```python
#investable universe
investable_universe = []

for col in df_full.columns:
    if ~df_full[col].isna().any():
        investable_universe.append(col)

len(investable_universe)
```

```python
df_investable = df_full[investable_universe]
df_investable
```

```python
def get_investable(stock, t):
    y = returns[[stock]].loc[:t, :]
    y.columns = ['Outcome']
    y = y.sort_index(ascending = False)
    
    X = returns.loc[:(t + pd.DateOffset(-1)), :]
    X = X.sort_index(ascending = False)
    
    df_full = pd.concat([X, y], axis = 1)
    df_full = df_full.sort_index(ascending = False)
    df_full['Outcome'] = df_full['Outcome'].shift(1)
    df_full = df_full.iloc[1:, :]
    
    investable_universe = []

    for col in df_full.columns:
        if ~df_full[col].isna().any():
            investable_universe.append(col)
            
    df_investable = df_full[investable_universe]
    
    return df_investable
```

```python
get_investable("921795", t)
```
