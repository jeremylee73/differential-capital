# Data Preparation

```python
#drop very first row
returns = returns.iloc[1:, :]
```

```python
returns
```

```python
#sample stock
stock = '905270'
```

```python
#time period
t = pd.to_datetime('1999-02-24')
```

```python
#save the target (sample stock's returns upto time t)
y = returns[[stock]].loc[:t, :]
y.columns = ['Outcome']
y = y.sort_index(ascending = False)
y
```

```python
#shift the features
X = returns.loc[:(t + pd.DateOffset(-1)), :]
X = X.sort_index(ascending = False)
X
```

```python
#concatenate outcome and features
df_full = pd.concat([X, y], axis = 1)
df_full = df_full.sort_index(ascending = False)
df_full
```

```python
#shit outcome down 1 row
df_full['Outcome'] = df_full['Outcome'].shift(1)
df_full = df_full.iloc[1:, :]
df_full
```

```python
#investable universe
investable_universe = []
last_days = 80

for col in df_full.columns:
    if ~df_full[col].iloc[:last_days].isna().any():
        investable_universe.append(col)

len(investable_universe)
```

```python
df_investable = df_full[investable_universe].iloc[:last_days]
```
